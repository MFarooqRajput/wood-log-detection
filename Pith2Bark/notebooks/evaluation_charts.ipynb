```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Log Identification
models = ['Grayscale', 'Grayscale', 'Grayscale', 'Grayscale', 'Grayscale', 'Polar Pith Canny(blur otsu)', 'Polar Pith Canny(blur triangle)', 'Polar Pith Canny(blur triangle)', 'Polar Canny(triangle)', 'Canny(triangle)']
baseline = [21.367, 32.206, 36.633, 57.035, 22.26]
winner = [12.829, 19.888, 13.429, 39.684, 18.79]

# sort both lists together
#baseline, models_baseline = zip(*sorted(zip(baseline, models), reverse=True))
#winner, models_winner = zip(*sorted(zip(winner, models), reverse=True))

n_groups = 5

fig, ax = plt.subplots(figsize = (10, 6))

#fig = plt.figure(figsize = (10, 6))

index = np.arange(n_groups)
bar_width = 0.4

baseline_bar = plt.bar(index - bar_width/2, baseline, bar_width, color = 'r', label = 'Baseline')
winner_bar = plt.bar(index + bar_width/2, winner,  bar_width, color = 'g', label = 'Winner')

plt.xticks(np.append(index - bar_width/2, index + bar_width/2), # tick positions
           models, # label corresponding to each tick
           rotation=45, horizontalalignment="right")

data = np.array([baseline, winner])
# add text annotations to each bar
for i, j in enumerate(data):
    for x_val, y_val in zip(index, j):
        ax.annotate(str(y_val), xy=(x_val+i*bar_width-bar_width/2, y_val), ha='center', va='bottom')

    
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.title('Models RMSE')
plt.legend()

plt.show()
```

```python
# Log Identification

N=5

models = ['Grayscale', 'Grayscale', 'Grayscale', 'Grayscale', 'Grayscale', 'Polar Pith Canny(blur otsu)', 'Polar Pith Canny(blur triangle)', 'Polar Pith Canny(blur triangle)', 'Polar Canny(triangle)', 'Canny(triangle)']
baseline = [21.367, 32.206, 36.633, 57.035, 22.26]
winner = [12.829, 19.888, 13.429, 39.684, 18.79]

ind = np.arange(N)
width=0.4

fig, ax = plt.subplots()
b1 = ax.bar(ind, baseline, width, edgecolor="w", linewidth=3, color = 'r', label = 'Baseline')
b2 = ax.bar(ind+width, winner, width, edgecolor="w",linewidth=3, color = 'g', label = 'Winner') 

#ax.set_xticks(ind+width/2)

ax.set_xticks(np.append(ind, ind + width), # tick positions
           models, # label corresponding to each tick
           rotation=45, horizontalalignment="right")

data = np.array([baseline, winner])
# add text annotations to each bar
for i, j in enumerate(data):
    for x_val, y_val in zip(index, j):
        ax.annotate(str(y_val), xy=(x_val+i*width, y_val), ha='center', va='bottom', fontsize=8)
        
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.title('Models RMSE')
plt.legend()
        
plt.show()
```
