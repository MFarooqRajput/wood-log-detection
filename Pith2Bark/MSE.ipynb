# MSE

```python
import warnings
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Disable warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('mse.csv', index_col='Name')
print(df)
```

```python
min_mse = df[["Inc/Dec", "Peaks", "High/Low", "Upward/Downward"]].min(axis=1).min()
max_mse = df[["Inc/Dec", "Peaks", "High/Low", "Upward/Downward"]].max(axis=1).max()
no_of_bins = max_mse // 5
print("Min: %s, Max: %s" % (min_mse, max_mse))
```

```python
df.plot(kind="bar")

plt.xticks(rotation=45, horizontalalignment="center")
plt.title("Model vs MSC")
plt.xlabel("Model")
plt.ylabel("MSE")
plt.axhline(y=min_mse, color='g', linestyle='--', linewidth=1)
plt.locator_params(axis='y', nbins=no_of_bins)
plt.show()
```
